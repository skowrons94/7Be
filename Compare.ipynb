```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exfor = "data/AZURE2/Abramovich-A0244002.dat"
ian = "data/IAN/Abramovich.dat"

data_exfor = np.loadtxt(exfor)
data_ian = np.loadtxt(ian)

plt.figure(figsize=(10, 6))

plt.errorbar( data_exfor[:, 0], data_exfor[:, 2], yerr=data_exfor[:, 3], fmt='o', label='EXFOR', color='blue')
plt.errorbar( data_ian[:, 0], data_ian[:, 2], yerr=data_ian[:, 3], fmt='o', label='IAN', color='red', markersize=1)

plt.xlabel('Energy (MeV)')
plt.ylabel('Cross Section (mb)')

plt.title('Comparison of EXFOR and IAN Data')
plt.legend()

plt.yscale('log')

plt.show()
```
